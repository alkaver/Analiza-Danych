# Opis metod

### Regresja Logistyczna

Jest to metoda statystyczna stosowana do modelowania zmiennej zależnej, która jest dychotomiczna (mająca tylko dwa możliwe wyniki, zazwyczaj kodowane jako 0 lub 1). Regresja logistyczna wykorzystuje funkcję logistyczną do oszacowania prawdopodobieństwa wystąpienia danego zdarzenia. Funkcja logistyczna, znana również jako funkcja sigmoidalna, jest definiowana jako:
<br>
$$
\sigma(z) = \frac{1}{1 + e^{-z}}
$$
gdzie  $z$  jest kombinacją liniową zmiennych niezależnych (cech) oraz ich odpowiednich współczynników:
$$
z = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n
$$
W modelu regresji logistycznej, celem jest oszacowanie współczynników $\beta$, które maksymalizują funkcję wiarygodności:
$$
L(\beta) = \prod_{i=1}^{n} \left( \sigma(z_i)^{y_i} \cdot (1 - \sigma(z_i))^{1 - y_i} \right)
$$
gdzie $y_i $ to obserwowane wartości zmiennej zależnej, a $ z_i $ to wartość funkcji liniowej dla i-tej obserwacji.

Model ten jest szczególnie przydatny, gdy zmienne niezależne mogą być ciągłe lub kategoryczne. Jego prostota interpretacji polega na tym, że współczynniki $\beta$ mogą być interpretowane jako logarytmy szans (log-odds), co ułatwia zrozumienie wpływu poszczególnych zmiennych na prawdopodobieństwo wystąpienia zdarzenia.

Kluczową zaletą regresji logistycznej jest jej zdolność do przewidywania prawdopodobieństw oraz łatwość w interpretacji współczynników jako wpływu zmiennych niezależnych na wynik. Ponadto, model ten jest efektywny w różnych problemach klasyfikacyjnych, takich jak medycyna, ekonomia czy nauki społeczne.

Cytowanie pracy: Cox, D. R. (1958). The regression analysis of binary sequences. Journal of the Royal Statistical Society: Series B (Methodological), 20(2), 215-232.

### Random Forrest


Jest to metoda zespołowa, która wykorzystuje wiele drzew decyzyjnych w celu klasyfikacji lub regresji. Losowość w budowie drzew i głosowanie większościowe pozwala na uniknięcie nadmiernego dopasowania om Forest**: (overfittingu), a elastyczność modelu umożliwia przetwarzanie dużych zbiorów danych i wydobycie istotnych cech.

Podstawowym elementem lasu losowego jest drzewo decyzyjne, które działa poprzez podział danych na mniejsze podzbiory na podstawie wartości atrybutów. Każde drzewo jest zbudowane na innym losowym podzbiorze danych treningowych, co zwiększa różnorodność modeli. Ważnym aspektem lasu losowego jest również losowy wybór podzbioru cech do oceny na każdym węźle drzewa, co dodatkowo zwiększa różnorodność i redukuje korelację między drzewami.

Ostateczna klasyfikacja lub przewidywanie w lesie losowym jest wynikiem głosowania większościowego wszystkich drzew w lesie:
$$
\hat{y} = \text{mode}(\{h_1(x), h_2(x), \dots, h_B(x)\})
$$
gdzie $h_{i}(x)$ to wynik i-tego drzewa decyzyjnego, a $B$ to liczba drzew w lesie.

Elastyczność lasu losowego sprawia, że jest on skuteczny w przetwarzaniu zarówno zmiennych ciągłych, jak i kategorycznych, oraz w radzeniu sobie z brakującymi danymi i dużymi zbiorami danych. Jest szeroko stosowany w dziedzinach takich jak bioinformatyka, finanse, marketing czy diagnostyka medyczna.

Cytowanie pracy: Breiman, L. (2001). Random forests. Machine learning, 45(1), 5-32.

### Parametry podane dla modelu Random Forest  <br> <br>

**n_estimators**  <br>
Opis: Liczba drzew decyzyjnych w lesie. <br>
Literatura: Zwiększenie liczby drzew zazwyczaj poprawia dokładność modelu, ponieważ zmniejsza wariancję modelu, co prowadzi do lepszego uogólnienia na nowe dane .  <br>
Wartości: [50, 100, 200] <br>
Uzasadnienie wyboru: Te wartości zostały wybrane, aby zbadać, jak liczba drzew wpływa na wydajność modelu. 50 drzew może być wystarczające dla prostych problemów, 100 to standardowa wartość startowa, a 200 to bardziej zaawansowana konfiguracja, która może poprawić dokładność kosztem większej złożoności obliczeniowej.
 <br> <br>**max_depth** <br>
Opis: Maksymalna głębokość drzewa decyzyjnego. <br>
Literatura: Ograniczenie głębokości drzewa pomaga w zapobieganiu nadmiernemu dopasowaniu modelu (overfitting) poprzez ograniczenie złożoności drzewa . <br>
Wartości: [None, 10, 20] <br>
Uzasadnienie wyboru: None pozwala drzewom rosnąć do pełnej głębokości, co może być przydatne dla bardzo złożonych danych. Głębokości 10 i 20 ograniczają złożoność modelu, co może poprawić jego zdolność generalizacji. <br> <br>**min_samples_split**<br>
Opis: Minimalna liczba próbek wymagana do podzielenia węzła. <br>
Literatura: Większe wartości tego parametru zapobiegają tworzeniu bardzo małych liści, co również może przeciwdziałać nadmiernemu dopasowaniu modelu . <br>
Wartości: [2, 5, 10] <br>

Uzasadnienie wyboru: 1 to wartość domyślna, umożliwiająca liściom zawierać pojedyncze próbki. 2 zwiększa minimalną liczbę próbek w liściu, co może prowadzić do bardziej stabilnych i ogólnych modeli. <br>  <br>

### Neural Network (Sieć Neuronowa)

Jest to zaawansowana metoda uczenia maszynowego inspirowana strukturą i działaniem ludzkiego mózgu. Składa się z warstw neuronów (węzłów), gdzie każda warstwa przetwarza dane wejściowe i przekazuje je do następnej warstwy. Sieci neuronowe są zdolne do wykrywania skomplikowanych zależności i wzorców w danych poprzez uczenie się i dostosowywanie wag połączeń między neuronami na podstawie danych treningowych.

Podstawowy element sieci neuronowej, neuron, oblicza ważoną sumę swoich wejść i przepuszcza ją przez funkcję aktywacji, np. funkcję sigmoidalną lub ReLU (Rectified Linear Unit):
$$
a = \sigma\left(\sum_{i=1}^{n} w_i x_i + b\right)
$$
gdzie $w_{i}$ to wagi połączeń, $x_{i}$ to wejścia, a $b$ to bias.

Sieci neuronowe składają się z jednej warstwy wejściowej, jednej lub więcej warstw ukrytych oraz warstwy wyjściowej. Proces uczenia polega na minimalizacji funkcji kosztu (np. błąd średniokwadratowy, entropia krzyżowa) przy użyciu algorytmu wstecznej propagacji (backpropagation) i optymalizatora (np. gradientu prostego, Adam):
$$
J(\theta) = \frac{1}{m} \sum_{i=1}^{m} L(y_i, \hat{y}_i)
$$
gdzie $L$ to funkcja kosztu, $y_i$ to prawdziwa wartość, a $\hat{y}_i$ to przewidywana wartość.

Elastyczność i moc obliczeniowa sieci neuronowych sprawiają, że są one idealne do rozwiązywania złożonych problemów klasyfikacyjnych i regresyjnych w dużych zbiorach danych. Są stosowane w wielu dziedzinach, takich jak rozpoznawanie obrazów, przetwarzanie języka naturalnego, prognozowanie i inne zaawansowane aplikacje.

Cytowanie pracy: McCulloch, W. S., & Pitts, W. (1943). A logical calculus of the ideas immanent in nervous activity. The bulletin of mathematical biophysics, 5(4), 115-133.<br>

## Parametry dla modelu sieci neuronowej 
**units_grid** <br> 
Opis: Liczba neuronów w każdej warstwie sieci neuronowej. <br>
Literatura: Liczba neuronów wpływa na zdolność modelu do uczenia się złożonych wzorców w danych. Większa liczba neuronów zwiększa zdolność modelu do reprezentowania skomplikowanych zależności, ale również może prowadzić do nadmiernego dopasowania (overfitting). <br>
Wartości: [(128, 64, 32), (256, 128, 64), (64, 32, 16)] <br>
Uzasadnienie wyboru: Te konfiguracje zostały wybrane, aby zbadać wpływ różnej liczby neuronów w warstwach na wydajność modelu. Małe konfiguracje (np. 64, 32, 16) mogą być wystarczające dla prostych problemów, natomiast większe konfiguracje (np. 256, 128, 64) mogą poprawić dokładność modelu kosztem większej złożoności obliczeniowej. <br>
Źródła:  <br>
Goodfellow, I., Bengio, Y., & Courville, A. (2016). Deep Learning. MIT Press. <br>
Géron, A. (2019). Hands-On Machine Learning with Scikit-Learn, Keras, and TensorFlow. O'Reilly Media. <br> <br>
**dropout_grid** <br> 
Opis: Dropout to technika regularizacji, w której losowo wyłączane są pewne neurony podczas treningu, aby zapobiec nadmiernemu dopasowaniu. <br>
Literatura: Dropout zmniejsza wariancję modelu i pomaga w zapobieganiu nadmiernemu dopasowaniu, co prowadzi do lepszego uogólnienia na nowe dane. <br>
Wartości: [0.3, 0.4, 0.5] <br>
Uzasadnienie wyboru: Te wartości są standardowo stosowane w praktyce i pozwalają na testowanie różnych poziomów regularizacji. 0.3 jest często używana jako początkowa wartość, 0.4 i 0.5 zwiększają stopień regularizacji, co może być korzystne w bardziej złożonych modelach. <br>
Źródła: <br>
Srivastava, N., Hinton, G., Krizhevsky, A., Sutskever, I., & Salakhutdinov, R. (2014). Dropout: A Simple Way to Prevent Neural Networks from Overfitting. Journal of Machine Learning Research, 15(1), 1929-1958. <br> <br> 
**batch_size_grid** <br> 
Opis: Liczba próbek używanych w jednej iteracji treningowej. <br>
Literatura: Wielkość batcha wpływa na stabilność i szybkość uczenia się modelu. Większe batchy mogą prowadzić do bardziej stabilnych gradientów, ale mniejsze batchy mogą wprowadzać więcej szumu, co może pomóc w ucieczce z lokalnych minimów. <br>
Wartości: [64, 128] <br>
Uzasadnienie wyboru: Batch size 64 jest często używana, ponieważ zapewnia dobry kompromis między stabilnością a szybkością uczenia. Batch size 128 jest większa i może prowadzić do bardziej stabilnych gradientów. <br>
Źródła: <br>
Keskar, N. S., Mudigere, D., Nocedal, J., Smelyanskiy, M., & Tang, P. T. P. (2016). On Large-Batch Training for Deep Learning: Generalization Gap and Sharp Minima. arXiv preprint arXiv:1609.04836. <br> <br> 
**learning_rate_grid** <br> 
Opis: Szybkość uczenia się kontroluje wielkość kroków, jakie model wykonuje w kierunku minimalizacji funkcji kosztu. <br>
Literatura: Wartość learning rate wpływa na szybkość konwergencji modelu. Zbyt duża wartość może prowadzić do niestabilności, a zbyt mała do bardzo wolnej konwergencji. <br>
Wartości: [0.001, 0.0001]  <br>
Uzasadnienie wyboru: 0.001 jest często stosowaną wartością początkową, która zapewnia szybkie uczenie się bez dużego ryzyka niestabilności. 0.0001 jest mniejszą wartością, stosowaną do bardziej precyzyjnego dostrajania modelu. <br>
Źródła: <br>
Bengio, Y. (2012). Practical Recommendations for Gradient-Based Training of Deep Architectures. In Neural Networks: Tricks of the Trade (pp. 437-478). Springer. <br> <br>
**layer_configurations** <br>
Opis: Konfiguracja liczby warstw w modelu sieci neuronowej. <br>
Literatura: Głębsze sieci mają większą zdolność do uczenia się bardziej skomplikowanych wzorców, ale mogą również prowadzić do problemów z gradientem i nadmiernym dopasowaniem. <br>
Wartości: [[3], [4], [5]] <br>
Uzasadnienie wyboru: Te konfiguracje pozwalają na testowanie różnych głębokości sieci, co jest kluczowe dla znalezienia optymalnej złożoności modelu w kontekście danego zadania. <br>
Źródła: <br> 
He, K., Zhang, X., Ren, S., & Sun, J. (2016). Deep Residual Learning for Image Recognition. In Proceedings of the IEEE Conference on Computer Vision and Pattern Recognition (pp. 770-778). <br> <br>
**epochs_grid** <br>
Opis: Liczba epok, czyli pełnych przejść przez cały zbiór treningowy. <br>
Literatura: Więcej epok zazwyczaj prowadzi do lepszego dopasowania modelu, ale może również prowadzić do nadmiernego dopasowania. <br>
Wartości: [5, 10, 15] <br>
Uzasadnienie wyboru: 5 epok może być wystarczające dla szybkiego prototypowania, 10 to standardowa wartość dla bardziej dokładnego trenowania, a 15 pozwala na jeszcze lepsze dopasowanie, ale kosztem większej liczby iteracji i potencjalnego ryzyka nadmiernego dopasowania. <br>

# Modeling: RandomForestClassifier

```python
from sklearn.model_selection import cross_val_score

param_values = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2]
}

best_accuracy = 0
best_params = None
best_rfc_model = None

for n_estimators in param_values['n_estimators']:
    for max_depth in param_values['max_depth']:
        for min_samples_split in param_values['min_samples_split']:
            for min_samples_leaf in param_values['min_samples_leaf']:
                clf = RandomForestClassifier(n_estimators=n_estimators,
                                             max_depth=max_depth,
                                             min_samples_split=min_samples_split,
                                             min_samples_leaf=min_samples_leaf,
                                             random_state=42)
                accuracies = cross_val_score(clf, X_train_scaled, y_train, cv=2)
                mean_accuracy = accuracies.mean()
                if mean_accuracy > best_accuracy:
                    best_accuracy = mean_accuracy
                    best_params = {
                        'n_estimators': n_estimators,
                        'max_depth': max_depth,
                        'min_samples_split': min_samples_split,
                        'min_samples_leaf': min_samples_leaf
                    }
                    best_rfc_model = clf


best_rfc_model.fit(X_train_scaled, y_train)
y_pred = best_rfc_model.predict(X_test_scaled)
accuracy = best_rfc_model.score(X_test_scaled, y_test)

rfc_pred_proba = best_rfc_model.predict_proba(X_test_scaled)[:, 1]

print("Najlepsza dokładność:", best_accuracy)
print("Najlepsze parametry:", best_params)
```

# Neural Network

```python
units_grid = [
    (128, 64, 32),
    (256, 128, 64),
    (64, 32, 16)
]
dropout_grid = [0.3, 0.4, 0.5]
batch_size_grid = [64, 128]
learning_rate_grid = [0.001, 0.0001]
layer_configurations = [
    [3],
    [4],
    [5] 
]
epochs_grid = [5, 10, 15] 

best_model = None
best_history = None
best_accuracy = 0


for units in units_grid:
    for dropout_rate in dropout_grid:
        for batch_size in batch_size_grid:
            for learning_rate in learning_rate_grid:
                for layers in layer_configurations:
                    for epochs in epochs_grid: 
                        num_layers = layers[0]


                        nn_model = Sequential()
                        nn_model.add(Input(shape=(X_train_scaled.shape[1],)))

                        for layer in range(num_layers):
                            if layer < len(units):
                                nn_model.add(Dense(units[layer], activation='relu'))
                            else:
                                nn_model.add(Dense(units[-1], activation='relu'))
                            nn_model.add(Dropout(dropout_rate))

                        nn_model.add(Dense(1, activation='sigmoid'))


                        nn_model.compile(tf.optimizers.Adam(learning_rate=learning_rate),
                                         loss='binary_crossentropy',
                                         metrics=['accuracy'])


                        history = nn_model.fit(X_train_scaled, y_train, epochs=epochs, batch_size=batch_size, validation_split=0.2, verbose=0)


                        val_accuracy = max(history.history['val_accuracy'])

                        if val_accuracy > best_accuracy:
                            best_accuracy = val_accuracy
                            best_model = nn_model
                            best_history = history
```

```python
nn_pred_proba = best_model.predict(X_test_scaled).flatten()
test_loss, test_accuracy = best_model.evaluate(X_test_scaled, y_test)
print('Dokładność testowa najlepszego modelu:', test_accuracy)


y_pred = (nn_pred_proba > 0.5).astype("int32")
print("Dokładność najlepszego modelu:", accuracy_score(y_test, y_pred))
```

```python
def evaluate_model(pred_proba, y_test, threshold=0.5):
    pred_class = (pred_proba >= threshold).astype(int)
    accuracy = accuracy_score(y_test, pred_class)
    report = classification_report(y_test, pred_class, output_dict=True)
    conf_matrix = confusion_matrix(y_test, pred_class)
    return accuracy, report, conf_matrix


lr_accuracy, lr_report, lr_conf_matrix = evaluate_model(lr_pred_proba, y_test)


rfc_accuracy, rfc_report, rfc_conf_matrix = evaluate_model(rfc_pred_proba, y_test)


nn_accuracy, nn_report, nn_conf_matrix = evaluate_model(nn_pred_proba, y_test)


hybrid_pred_proba = (weights[0] * lr_pred_proba) + (weights[1] * rfc_pred_proba) + (weights[2] * nn_pred_proba)
hybrid_accuracy, hybrid_report, hybrid_conf_matrix = evaluate_model(hybrid_pred_proba, y_test)


models = ['Logistic Regression', 'RandomForestClassifier', 'Neural Network', 'Hybrid Model']
accuracies = [lr_accuracy, rfc_accuracy, nn_accuracy, hybrid_accuracy]


plt.figure(figsize=(8, 6))
sns.heatmap(hybrid_conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Predicted Negative', 'Predicted Positive'],
            yticklabels=['True Negative', 'True Positive'])
plt.title('Macierz Pomyłek Dla Modelu Hybrydowego')
plt.xlabel('Przewidziane')
plt.ylabel('Rzeczywiste')
plt.show()


metrics = ['precision', 'recall', 'f1-score']
classes = list(hybrid_report.keys())[:-3]  


plot_data = {
    'Logistic Regression': {metric: [lr_report[cls][metric] for cls in classes] for metric in metrics},
    'RandomForestClassifier': {metric: [rfc_report[cls][metric] for cls in classes] for metric in metrics},
    'Neural Network': {metric: [nn_report[cls][metric] for cls in classes] for metric in metrics},
    'Hybrid Model': {metric: [hybrid_report[cls][metric] for cls in classes] for metric in metrics}
}

x = np.arange(len(classes))
width = 0.2


fig, ax = plt.subplots(2, 1, figsize=(12, 12))


ax[0].bar(models, accuracies, color=['blue', 'green', 'red', 'purple'])
ax[0].set_title('Dokładność Modeli')
ax[0].set_ylabel('Dokładność')
ax[0].set_ylim(0, 1)


for i, (model, data) in enumerate(plot_data.items()):
    for j, metric in enumerate(metrics):
        ax[1].bar(x + (i * width) - width, data[metric], width, label=f'{model} - {metric}' if i == 0 else "")

ax[1].set_xlabel('Klasy')
ax[1].set_ylabel('Wyniki')
ax[1].set_title('Raport Klasyfikacji Na Model')
ax[1].set_xticks(x)
ax[1].set_xticklabels(classes)
ax[1].legend()

fig.tight_layout()
plt.show()
```
